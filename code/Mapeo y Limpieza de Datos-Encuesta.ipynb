```python
import pandas as pd

# Cargar el conjunto de datos en un DataFrame
df = pd.read_csv(r"C:\DESERCIÓN ESCOLAR EN EMS (Respuestas) - Respuestas de formulario 1.csv")

# Mostrar las primeras filas para verificar las columnas y algunos valores
print(df.head())

# Mostrar solo los nombres de las columnas
print("\nNombres de las columnas:")
print(df.columns)
```

```python
# Descripción general de las variables numéricas
print("\nDescripción estadística de variables numéricas:")
print(df.describe())

# Tipo de datos de cada variable
print("\nTipo de datos de cada columna:")
print(df.dtypes)

# Valores únicos en variables categóricas
print("\nValores únicos por variable categórica:")
for column in df.select_dtypes(include='object').columns:
    print(f"{column}: {df[column].unique()}")
```

```python
# Identificar columnas con valores faltantes y decidir cómo manejarlos
print("\nValores faltantes por columna:")
print(df.isnull().sum())
```

```python
df = df.dropna()
df = df.dropna(axis=1)
```

```python
# Eliminar duplicados
df.drop_duplicates(inplace=True)
```

```python
# Convertir a una escala estandarizada
categorias_binarias = {
    '¿Hay situaciones familiares, económicas que dificultan tu permanencia en la preparatoria?': {'Si': 1, 'No': 0},
    '¿Crees que te haga falta apoyo psicológico o emocional para continuar y terminar tus estudios?': {'Si': 1, 'No': 0},
    '¿ Consideras que tienes acceso a los recursos y herramientas necesarias para estudiar?': {'Si': 1, 'No': 0},
    '¿Sientes que lo que aprendes en la preparatoria es relevante para tus metas personales?': {'Si': 1, 'No': 0},
    '¿Crees que la carga académica o el ritmo de estudios es demasiado intenso y puede desmotivarte a seguir estudiando?': {'Si': 1, 'No': 0},
    '¿Cómo influye tu círculo social (amigos, compañeros de clase, etc), en tu decisión de seguir estudiando?': {'Mucho': 2, 'Poco': 1, 'Nada':0},
    '¿Sientes que tienes tiempo suficiente para equilibrar tus estudios con otras responsabilidades?': {'Si': 1, 'No': 0},
    '¿Alguna vez has considerado que sería más beneficioso empezar a trabajar en lugar de estudiar?': {'Si': 1, 'No': 0}   
}
for col, mapping in categorias_binarias.items():
    if col in df.columns:
        df[col] = df[col].map(mapping)

# Eliminar duplicados en el conjunto de datos
df.drop_duplicates(inplace=True)

# Clasificar variables en categorías
socioeconomico = ['¿Hay situaciones familiares, económicas que dificultan tu permanencia en la preparatoria?', ' ¿Consideras que tienes acceso a los recursos y herramientas necesarias para estudiar?', '¿Alguna vez has considerado que sería más beneficioso empezar a trabajar en lugar de estudiar? ']  
rendimientoescolar = ['¿Crees que la carga académica o el ritmo de estudios es demasiado intenso y puede desmotivarte a seguir estudiando?', '¿Sientes que tienes tiempo suficiente para equilibrar tus estudios con otras responsabilidades?', '¿Cuáles serían los cambios que te ayudarían a sentirte mas motivad@ para continuar en la escuela?'] 
psicologico = ['¿Crees que te haga falta apoyo psicológico o emocional para continuar y terminar tus estudios?', ' ¿Sientes que lo que aprendes en la preparatoria es relevante para tus metas personales?', '¿Consideras que hay otro factor no mencionado que afecta tu permanencia escolar?']  

print("\nCategorías de variables:")
print(f"socioeconomico: {socioeconomico}")
print(f"rendimientoescolar: {rendimientoescolar}")
print(f"psicologico: {psicologico}")

# Resumen estadístico final del conjunto de datos limpio
print("\nResumen estadístico final del conjunto de datos limpio:")
print(df.describe())
```
